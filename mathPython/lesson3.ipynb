```python
import numpy as np 
import matplotlib.pyplot as plt 

a = 1.5
x = np.linspace(-1,1)
y = a * x

plt.plot(x, y)
plt.xlabel("x", size= 14)
plt.ylabel("y", size= 14)

plt.grid()
plt.show()
```
