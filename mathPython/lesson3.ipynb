```python
import numpy as np 
import matplotlib.pyplot as plt 

a = 1.5
x = np.linspace(-1,1)
y = a * x

plt.plot(x, y)
plt.xlabel("x", size= 14)
plt.ylabel("y", size= 14)

plt.grid()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def my_sin(x):
    return np.sin(x)

def my_cos(x):
    return np.cos(x)

x = np.linspace(-np.pi, np.pi)
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label = "sin")
plt.plot(x, y_cos, label = "cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)

plt.grid()
plt.show()
```
