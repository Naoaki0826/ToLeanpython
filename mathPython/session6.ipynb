```python
import numpy as np 
import matplotlib.pyplot as plt 

x = []
y = []
total = 0
num_5 = 0
n = 5000

for i in range (n):
    if np.random.randint(6)+1 == 5:
        num_5 += 1
    total += 1
    x.append(i)
    y.append(num_5/total)

plt.plot(x, y)
plt.plot(x, [1/6]*n, linestyle="dashed")
plt.xlabel("x", size=14)
plt.xlabel("y", size=14)

plt.grid()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

# 確率密度関数
def pdf(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2 / (2 * sigma**2))

x = np.linspace(-5 , 5)
y_1 = pdf(x, 0.0, 0.5)
y_2 = pdf(x, 0.0, 1.0)
y_3 = pdf(x, 0.0, 2.0)


plt.plot(x, y_1, label="σ: 0.5", linestyle = "dashed")
plt.plot(x, y_2, label="σ: 1.0", linestyle = "solid")
plt.plot(x, y_3, label="σ: 2.0", linestyle = "dashdot")

plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def power_func(x, c, k): 
    return c*x**(-k)

x = np.linspace(1, 5)
y_1 = power_func(x, 1.0, 1.0)
y_2 = power_func(x, 1.0, 2.0)
y_3 = power_func(x, 1.0, 4.0)


plt.plot(x, y_1, label="k: 0.5", linestyle = "dashed")
plt.plot(x, y_2, label="k: 2.0", linestyle = "solid")
plt.plot(x, y_3, label="k: 4.0", linestyle = "dashdot")

plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.array([50, 70, 40, 60, 80])
y = np.array([60, 80, 50, 50, 70])
z = np.array([60, 40, 60, 40, 30])

cov_xy = np.average( (x-np.average(x)) * (y - np.average(y)) )
print("cov_xy = ",cov_xy)

cov_xz = np.average( (x-np.average(x)) * (z - np.average(z)) )
print("cov_xy = ",cov_xz)

plt.scatter(x, y, marker="o", label="xy", s=40)
plt.scatter(x, z, marker="x", label="xz", s=60)

plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def show_cov(cov):
    print("---  Covatiance:", cov, " ---")
    average = np.array([0 ,0])
    cov_matrix = np.array([[1, cov], [cov, 1]]) #共分散を行列で指定

    data = np.random.multivariate_normal(average, cov_matrix, 3000)

    x = data[:, 0] #最初の列をx座標に
    y = data[:, 1] #次の列をy座標に
    
    plt.scatter(x, y, marker="x", s = 20)
    plt.xlabel("x", size = 14)
    plt.ylabel("y", size = 14)
    plt.grid()

    plt.show()


show_cov(0.6)
show_cov(0.0)
show_cov(-0.6)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x_data = np.array([2.4, 1.2, 3.5, 2.1, 4.7])
y_data = np.zeros(5)

mu = np.average(x_data) #平均値
print(mu)
sigma = np.std(x_data) #標準偏差
print(sigma)


def pdf(x, mu, sigma): # 確率密度関数
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))

def log_likelinhood(p):
    return np.sum(np.log(p)) # 対数尤度

x_sigma = np.linspace(0.5, 8)
y_log_like = [] # 縦軸に使う対数尤度

for s in x_sigma:
    log_like = log_likelinhood(pdf(x_data, mu, s))
    y_log_like.append(log_like) # 対数尤度を縦軸に追加

plt.plot(x_sigma, np.array(y_log_like))
plt.plot([sigma, sigma], [min(y_log_like),max(y_log_like)], linestyle ="dashed")


plt.xlabel("x_sigma", size=14)
plt.ylabel("y_sigma", size=14)
plt.grid()

plt.show()
```

```python
import numpy as np 

delta = 1e-7

def cross_entropy(p, t):
    return -np.sum(t*np.log(p+delta) + (1-t)*np.log(1-p+delta))

p_1 = np.array([0.2, 0.8, 0.1, 0.3, 0.9, 0.7]) #正解から離れている
p_2 = np.array([0.7, 0.3, 0.9, 0.8, 0.1, 0.2]) #正解に近い
t   = np.array([1, 0, 1, 1, 0, 0])

print("--- 予測と正解が離れている ---")
print(cross_entropy(p_1, t))


print("--- 予測と正解が近い ---")
print(cross_entropy(p_2, t))
```
