```python
import numpy as np 
import matplotlib.pyplot as plt 

def my_func(x):
    return 3*x**2 + 4*x - 5

def my_func_dif(x):
    return 6*x + 4

x = np.linspace(-3, 3)
y = my_func(x)

a = 1
y_t = my_func_dif(a)*x + my_func(a) - my_func_dif(a) * a

plt.plot(x, y, label = "y")
plt.plot(x, y_t, label = "y_t")

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)

plt.show()


```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def my_func(x):
    return x**2 - 2*x

def grad_func(x):
    return 2*x - 2

eta = 0.1 #学習係数
x = 4.0 #xの初期値
record_x = []
record_y = []

for i in range(20):
    y = my_func(x)
    record_x.append(x)
    record_y.append(y)
    x -= eta * grad_func(x)

x_f = np.linspace(-2, 4) #表示範囲
y_f = my_func(x_f)


plt.plot(x_f, y_f, linestyle="dashed")
plt.scatter(record_x, record_y)

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)

plt.grid()

plt.show()
```

```python
import numpy  as np 
import matplotlib.pyplot as plt 

def my_func(x):
    return x**4+2*x**3-3*x**2-2*x

def grad_func(x):
    return 4*x**3+6*x**2-6*x-2

eta = 0.01
x = -0.5
record_x = []
record_y = []

for i in range(200):
    y = my_func(x)
    record_x.append(x)
    record_y.append(y)
    x -= eta * grad_func(x)

x_f = np.linspace(-2.8, 1.6)
y_f = my_func(x_f)

plt.plot(x_f, y_f, linestyle="dashed")
plt.scatter(record_x,record_y)

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)

plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

def my_func(x):
    return x**4 - 2*x**3 - 3*x**2 + 2*x

def grad_func(x):
    return 4*x**3 - 6*x**2 - 6*x + 2

eta = 0.01
x = 0.5

record_x = []
record_y = []

for i in range(200):
    y = my_func(x)
    record_x.append(x)
    record_y.append(y)

    x -= eta * grad_func(x)

x_f = np.linspace(-1.6, 2.8)
y_f = my_func(x_f)

plt.plot(x_f, y_f, linestyle="dashed")
plt.scatter(record_x,record_y)

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)

plt.grid()

plt.show()
```
